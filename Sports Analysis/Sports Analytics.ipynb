#### Converting Value and Wage (string data type) to float data type---

```python
# convert string of "Value" & "Wages" into float value
def str2float(euros):
    # convert m into millions
    if euros[-1] == "M":
        return float(euros[1:-1])*1000000
    elif euros[-1] == "K":
        return float(euros[1:-1])*1000
    else:
        return float(euros[1:])

fifa_df["Value"] = fifa_df["Value"].apply(lambda x : str2float(x))
fifa_df["Wage"] = fifa_df["Wage"].apply(lambda x : str2float(x))

```

```python
fifa_df[['Name','Value','Wage']]
```

#### Joined----

```python
fifa_df['Joined'].isna().sum()
```

```python
fifa_df['Joined']=fifa_df['Joined'].fillna('0')
```

```python
def str4(x):
    if isinstance(x,str):
        return int(x[-4:])
```

```python
fifa_df['Joined'] = fifa_df['Joined'].apply(str4)
fifa_df['Joined'].head(5)
```

#### Contract Valid until------

```python
fifa_df['Contract Valid Until']=pd.to_datetime(fifa_df['Contract Valid Until'])
fifa_df['Contract Valid Until']
```

#### Height-----

```python
def feet(x):
    l=[]
    if isinstance(x,str):
        l = x.split("'")
        i = int(l[0])
        j = int(l[1])
        f = ((i*12)+j)/12
        return f
fifa_df['Height'] = fifa_df['Height'].apply(feet) 
```

```python
fifa_df['Height'].head()
```

#### Weight-----

```python
def llbs(x):
    l=[]
    if isinstance(x,str):
        l = x.split("lbs")
        i = l[0]
        return float(i)
fifa_df['Weight'] = fifa_df['Weight'].apply(llbs)
```
